#### matrix transpose

`````python
matrix_a = np.asarray([
    [0.7, 3],
    [1.7, 2],
    [0.7, 9]
], dtype=np.float32)

matrix_b = np.asarray([
    [113, 3, 10],
    [1, 0, 1],
], dtype=np.float32)

transpose_a = np.transpose(matrix_a)
print('A^T\n', transpose_a)
matrix_a = np.transpose(transpose_a)
print('\n(A^T)^T\n', matrix_a)

trans_ba = np.dot(np.transpose(matrix_b), transpose_a)    # B^T * A^T
trans_ab = np.dot(transpose_a, np.transpose(matrix_b))    # A^T * B^T
print('\nB^T * A^T\n', trans_ba)

product_ab = np.dot(matrix_a, matrix_b)                   # (AB)^T
print('\n(AB)^T\n', np.transpose(product_ab))
`````

`````python
matrix_a = np.asarray([
    [1.5, 3],
    [1, 4]
])
def matrix_inverse_two(mat):
    det = (mat[0,0]*mat[1,1] - mat[0,1]*mat[1,0])  # determinant
    if det == 0:
        raise ValueError("The matrix isn't invertible")
    right_mat = np.asarray([
        [mat[1,1], -mat[0,1]],
        [-mat[1,0], mat[0,0]]
    ])
    inv_mat = np.dot(1/det, right_mat)
    return inv_mat
inverse_a = matrix_inverse_two(matrix_a)
i_2 = np.dot(inverse_a, matrix_a)
print(i_2)
`````

#### gradient descent
```single parameter gradient descent
   cost function / the loss function
   MSE(a1) = 1/n[SUM(a1*x(i) - y(i))^2
   dMSE(a1)/da1 = (2/n)SUM[x(i) * (a1*x(i) -y(i))]
   a1 := a1 - Alpha* dMSE(a1)/da1   alpha=learning rate```

`````python
def derivative(a1, xi_list, yi_list):
    len_data = len(xi_list)
    error = 0
    for i in range(0, len_data):
        error += xi_list[i]*(a1*xi_list[i] - yi_list[i])
    deriv = 2*error/len_data
    return deriv

def gradient_descent(xi_list, yi_list, max_iterations, alpha, a1_initial):
    a1_list = [a1_initial]

    for i in range(0, max_iterations):
        a1 = a1_list[i]
        deriv = derivative(a1, xi_list, yi_list)
        a1_new = a1 - alpha*deriv
        a1_list.append(a1_new)
    return(a1_list)

param_iterations = gradient_descent(train['Gr Liv Area'], train['SalePrice'], 20, .0000003, 150)
final_param = param_iterations[-1]
print(final_param)
`````

```multi parameter gradient descent
   cost function / the loss function
   MSE(a0, a1) = (1/2n)[SUM(a0 + a1*x(i) - y(i))^2
   dMSE/da0 = (1/n)SUM[(a0 + a1*x(i) -y(i))]
   dMSE/da1 = (1/n)SUM[x1(i) * (a0 + a1*x(i) -y(i))]
   dMSE/dan = (1/n)SUM[xn(i) * (a0 + an*x(i) -y(i))]
   
   a1 := a1 - Alpha* dMSE(a0,a1)/da1   alpha=learning rate
   a0 := a0 - Alpha* dMSE(a0,a1)/da0   alpha=learning rate
```

`````python
def a1_derivative(a0, a1, xi_list, yi_list):
    len_data = len(xi_list)
    error = 0
    for i in range(0, len_data):
        error += xi_list[i]*(a0 + a1*xi_list[i] - yi_list[i])
    deriv = error/len_data
    return deriv

def a0_derivative(a0, a1, xi_list, yi_list):
    len_data = len(xi_list)
    error = 0
    for i in range(0, len_data):
        error += a0 + a1*xi_list[i] - yi_list[i]
    deriv = error/len_data
    return deriv

def gradient_descent(xi_list, yi_list, max_iterations, alpha, a1_initial, a0_initial):
    a1_list = [a1_initial]
    a0_list = [a0_initial]

    for i in range(0, max_iterations):
        a1 = a1_list[i]
        a0 = a0_list[i]
        a1_deriv = a1_derivative(a0, a1, xi_list, yi_list)
        a0_deriv = a0_derivative(a0, a1, xi_list, yi_list)
        a1_new = a1 - alpha*a1_deriv
        a0_new = a0 - alpha*a0_deriv
        a1_list.append(a1_new)
        a0_list.append(a0_new)
    return(a0_list, a1_list)

a0_params, a1_params = gradient_descent(train['Gr Liv Area'], train['SalePrice'], 20, .0000003, 150, 1000)
print(a0_params[-1], '  ', a1_params[-1])
`````

#### ordinary least squares

``` OLS estimation provides a closed form solution to the problem of finding the optimal parameter values. 
    A closed form solution is one where a solution can be computed arithmetically with a predictable amount of mathematical operations.
 Cost Function: J(a) = (1/n) (X*a-y)^T * (X*a-y)
 dJ(a)/da = 2(X^T)Xa -2(X^T)y
 solving a = ((X^T)X)^(-1)*(X^T)*y
 computing a matrix inverse has a computational complexity of apprximately O(n^3) ```
